```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.normal(size=1000)
```

```python
plt.hist(x);
```

```python
plt.hist(x, density=True);
```

```python
plt.hist(x, density=True, bins=np.linspace(-5,5,21));
```

```python
x = np.random.gamma(2,3,100000)
```

```python
plt.hist(x, bins = 30);
```

```python
plt.hist(x, bins = 30, density=True);
```

```python
plt.hist(x, bins=30, cumulative=30);
```

```python
plt.hist(x, bins=30, density=True, cumulative=30, histtype="step");
```

```python
plt.figure()
plt.subplot(221)
np.linspace(-5,5,21)
plt.subplot(222)
plt.hist(x, bins=30, density=True);
plt.subplot(223)
plt.hist(x, bins=30, cumulative=30);
plt.subplot(224)
plt.hist(x, bins=30, density=True, cumulative=30, histtype="step");
```
