```python
import matplotlib.pyplot as plt
import numpy as np
import random
import time
```

```python
delta_x = np.random.normal(0,1,(2,5))
```

```python
plt.plot(delta_x[0], delta_x[1], "go");
```

```python
x = np.cumsum(delta_x, axis=1)
```

```python
delta_x = np.random.normal(0,1,(2,5))
x = np.cumsum(delta_x, axis=1)
plt.plot(x[0], x[1], "ro-");
plt.savefig("rw.pdf")
```

```python
x_0 = np.array([[0],[0]])
delta_x = np.random.normal(0,1,(2,5))
x = np.concatenate((x_0, np.cumsum(delta_x, axis=1)), axis=1)
```

```python
x_0 = np.array([[0],[0]])
delta_x = np.random.normal(0,1,(2,5))
x = np.concatenate((x_0, np.cumsum(delta_x, axis=1)), axis=1)
plt.plot(x[0], x[1], "ro-");
plt.savefig("rw2.pdf")
```

```python
x_0 = np.array([[0],[0]])
delta_x = np.random.normal(0,1,(2,100))
x = np.concatenate((x_0, np.cumsum(delta_x, axis=1)), axis=1)
plt.plot(x[0], x[1], "ro-");
plt.savefig("rw3.pdf")
```

```python
# How are the displacements, the individual steps, in the random walk related to one another?
# Any two consecutive displacements are independent of one another.
```
